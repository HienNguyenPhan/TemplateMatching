```python
import numpy as np
from skimage.io import imshow, imread
from skimage.color import rgb2gray
from skimage.feature import canny
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load and convert to grayscale
sample = imread('rabbit.jpeg')
sample_g = rgb2gray(sample)

# Apply Canny edge detection to the grayscale image
edges_full = canny(sample_g, sigma=1)

# Display original and grayscale images
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(sample)
ax[1].imshow(sample_g, cmap='gray')
ax[0].set_title('Colored Image', fontsize=15)
ax[1].set_title('Grayscale Image', fontsize=15)
plt.show()

# Define coordinates for the patch
left_x, left_y = 110, 570
right_x, right_y = 210, 730

# Extract the patch and apply Canny to the patch
patch = sample_g[left_y:right_y, left_x:right_x]
edges_patch = canny(patch, sigma=1)

# Show patch location on the full image and the patch itself with edges
fig, ax = plt.subplots(1, 3, figsize=(15, 10))
ax[0].imshow(edges_full, cmap='gray')
ax[0].add_patch(Rectangle((left_x, left_y), right_x-left_x, right_y-left_y, edgecolor='r', facecolor='none'))
ax[0].set_title('Canny Edges of Full Image', fontsize=15)

# Show the patch location
ax[1].imshow(patch, cmap='gray')
ax[1].set_title('Original Patch', fontsize=15)

# Show Canny edges of the patch
ax[2].imshow(edges_patch, cmap='gray')
ax[2].set_title('Canny Edges of Patch', fontsize=15)

plt.show()
```
